```python
import torch
import numpy as np
from torchvision.transforms import Compose, ToTensor, Normalize, Lambda, GaussianBlur
import matplotlib.pyplot as plt
```

```python
a = torch.rand(1,28,28)


transform = Compose([
        GaussianBlur(5, 7),
        GaussianBlur(7, 14),
        GaussianBlur(5, 7),
        Lambda(lambda x: x > 0.5)])

b = transform(a)

plt.imshow(a[0])
plt.show()
plt.imshow(b[0])
plt.show()
```
